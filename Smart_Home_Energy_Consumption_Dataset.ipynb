# DataSet Link : https://www.kaggle.com/datasets/taranvee/smart-home-dataset-with-weather-information

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("HomeC.csv")
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Value Map")
plt.show()
```

```python

df['time'] = pd.to_numeric(df['time'], errors='coerce')
df = df.dropna(subset=['time'])
df['time'] = df['time'].astype(int)
df['time'] = pd.to_datetime(df['time'], unit='s')
df.set_index('time', inplace=True)
```

```python
df['hour'] = df.index.hour
df['day'] = df.index.day
df['weekday'] = df.index.weekday  # 0 = Monday
df['month'] = df.index.month
df['is_weekend'] = (df['weekday'] >= 5).astype(int)
```

```python
plt.figure(figsize=(12,5))
df['use [kW]'].plot()
plt.title("Total Household Energy Usage Over Time")
plt.ylabel("kW")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['temperature'], y=df['use [kW]'], s=10)
plt.title("Temperature vs Energy Consumption")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Energy (kW)")
plt.show()
```

```python
appliance_cols = ['Dishwasher [kW]', 'Fridge [kW]', 'Microwave [kW]',
                  'Living room [kW]', 'Kitchen 12 [kW]', 'Barn [kW]']

df[appliance_cols].mean().sort_values().plot(kind='bar', figsize=(10,5))
plt.title("Average Appliance Energy Consumption")
plt.ylabel("kW")
plt.show()
```

```python
weather_cols = ['temperature','humidity','visibility','apparentTemperature',
                'pressure','windSpeed','dewPoint','precipIntensity']

energy_cols = ['use [kW]', 'gen [kW]', 'House overall [kW]']

corr = df[weather_cols + energy_cols].corr()

plt.figure(figsize=(12,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Weather vs Energy Correlation")
plt.show()
```

```python
df['hour'] = df.index.hour

plt.figure(figsize=(10,5))
sns.lineplot(x='hour', y='use [kW]', data=df, estimator='mean')
plt.title("Average Energy Usage by Hour of Day")
plt.ylabel("kW")
plt.show()
```

```python
df['weekday'] = df.index.weekday

plt.figure(figsize=(10,5))
sns.boxplot(x='weekday', y='use [kW]', data=df)
plt.title("Energy Usage by Day of Week")
plt.xlabel("Weekday (0=Mon)")
plt.ylabel("kW")
plt.show()
```
