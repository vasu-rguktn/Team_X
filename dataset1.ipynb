```python
import pandas as pd

file_path = "/content/household_power_consumption.txt"  # update if in a folder

# Load the dataset
df = pd.read_csv(file_path, sep=';', low_memory=False)

# See the first few rows
df.head()
```

```python
df.replace('?', pd.NA, inplace=True)

# Step 3: Convert numeric columns to float
numeric_cols = df.columns[2:]  # Usually, first two columns are Date and Time
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Converts invalid to NaN

# Step 4: Check result
print(df.info())
print(df.head())

# Step 5: Example numeric operation (sum of Global_active_power)
print("Sum of Global Active Power:", df['Global_active_power'].sum(skipna=True))
```
