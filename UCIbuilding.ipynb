```python
import matplotlib.pyplot as plt
plt.hist(data['Appliances'],bins=50,color='skyblue')
plt.title("Distribution of Appliances Energy Usage")
plt.xlabel("Energy Consumption")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(data['T_out'][:200],label='Outdoor Temp')
plt.plot(data['T1'][:200],label='Indooe Temp Sensor 1')
plt.legend()
plt.title("Temperature Sensors (First 200 Records)")
plt.show()
```
